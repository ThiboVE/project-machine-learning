```python
import numpy as np
import torch
from library.GCN import *
from pathlib import Path
```

```python
main_path = Path.cwd().parents[0]
data_path = main_path / "data" / "RDKit" / "rdkit_only_valid_smiles_qm9.pkl"

df = pd.read_pickle(data_path)
smiles_list = df['SMILES']
```

```python
# Check the highest atomic number present in all molecules
max_atom_num = 0
max_num_atoms = 0
max_atom_charge = 0

for smiles in smiles_list:

    mol = Chem.MolFromSmiles(smiles)
    mol = Chem.AddHs(mol)
    atoms = mol.GetAtoms()
    
    num_atoms = len(list(atoms))
    if num_atoms > max_num_atoms:
        max_num_atoms = num_atoms

    for atom in atoms:
        atom_num = atom.GetAtomicNum()
        atom_charge = atom.GetFormalCharge()

        if atom_num > max_atom_num:
            max_atom_num = atom_num
        if atom_charge > max_atom_charge:
            max_atom_charge = atom_charge
```

```python
print(f'Highest atomic number: {max_atom_num}')
print(f'Highest number of atoms in a molecule: {max_num_atoms} (with Hs)')
print(f'Highest formal charge: {max_atom_charge}')
```
